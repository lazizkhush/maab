```python
import pandas as pd
import sqlite3

with sqlite3.connect('data/chinook.db') as connection:
    customers_df = pd.read_sql("Select * From Customers", con=connection)
    invoices_df = pd.read_sql("Select * from Invoices", con=connection)
merged = pd.merge(customers_df, invoices_df, on='CustomerId')
invoices = invoices_df['CustomerId'].value_counts().reset_index()
print(invoices)
```

```python
df_titanic = pd.read_excel('data/titanic.xlsx')
avg_age = df_titanic.groupby('Pclass')['Age'].mean()
total_fare = df_titanic.groupby('Pclass')['Fare'].sum()
count_of_passengers = df_titanic.groupby('Pclass')['Pclass'].count()

result = pd.DataFrame({
    "Average_age": avg_age,
    'Total_fare': total_fare,
    'Number_of_passengers' : count_of_passengers
}).reset_index()

print(result)
```

```python
df_titanic = pd.read_excel('data/titanic.xlsx')
def categorize(x):
    if x>18:
        return 'Adult'
    return 'Child'
ct = df_titanic['Age'].apply(categorize)
df_titanic['Age_group'] = ct
print(df_titanic)
```

```python
df_employee = pd.read_csv('data/employee.csv')
df_employee['Normalized Salary'] = df_employee.groupby('DEPARTMENT')['BASE_SALARY'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)
print(df_employee)
```

```python
def categorize_duration(x):
    if x<60:
        return 'short'
    elif x>=60 and x<=120:
        return 'medium'
    return 'long'

df_movie = pd.read_csv('data/movie.csv')
df_movie['duration_category'] = df_movie['duration'].apply(categorize_duration)
print(df_movie)
```

```python

def filter_survivors(df):
    return df[df['Survived'] == 1]

def fill_missing_age(df):
    mean_age = df['Age'].mean()
    return df.assign(Age=df['Age'].fillna(mean_age))

def add_fare_per_age(df):
    df['Fare_Per_Age'] = df['Fare']/df['Age']
    return df

df_titanic_final = (
    df_titanic.pipe(filter_survivors)
    .pipe(fill_missing_age)
    .pipe(add_fare_per_age)
    )

print(df_titanic_final)
```

```python
def filter_delay(df):
    return df[df['DepDelay']>30]

def add_delay_per_hour(df):
    df['Delay_Per_Hour'] = df['DepDelay']/df['Duration']
    return df

df_flights_modified = (
    df_flights.pipe(filter_delay)
    .pipe(add_delay_per_hour)
    )

print(df_flights_modified)
```
