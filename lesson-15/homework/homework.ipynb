```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10,10)
y = x**2-4*x+4
plt.plot(y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 30)
sinx = np.sin(x)
cosx = np.cos(x)

plt.plot(sinx, color='r', marker='o', label='sin(x)')
plt.plot(cosx, color='g', linestyle=':', marker='x', label='cos(x)')
plt.legend()
plt.show()
```

```python
x = np.arange(1, 10, 0.1)

plt.subplot(2, 2, (1, 1))
plt.title("Cubic")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.plot(x**3)

plt.subplot(2, 2, 2)
plt.title('Sine')
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.plot(np.sin(x), c='r')

plt.subplot(2, 2, 3)
plt.title("e^x")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.plot(np.e**x, c='k')

plt.subplot(2, 2, 4)
plt.title('Logarithmic')
plt.plot(np.log(x+1), c='b')
plt.subplots_adjust(right=1.2, hspace=0.5)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
```

```python
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)


plt.title('Random 100 values')
plt.xlabel('x-axix')
plt.ylabel('y-label')
plt.grid(True)
plt.scatter(x, y, c='red', marker='x')
```

```python
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30, edgecolor='k', alpha=0.6)
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
ax = plt.axes(projection='3d')
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

xx, yy = np.meshgrid(x, y)

z = np.cos(xx**2+yy**2)

ax.plot_surface(xx, yy, z, cmap='viridis')
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
sales = [200, 150, 250, 175, 225]
plt.bar(categories, sales, width=0.6, color=['red', "blue", 'yellow', 'green', 'purple'])
plt.title('Sales data')
plt.xlabel('Products')
plt.ylabel('Sales')
```

```python
# Time periods
time_periods = ['T1', 'T2', 'T3', 'T4']

# Sample data for each category
category_A = [4, 7, 3, 6]
category_B = [5, 2, 6, 4]
category_C = [3, 5, 4, 7]

# Set figure size
plt.figure(figsize=(8, 6))

# Create stacked bar chart
plt.bar(time_periods, category_A, color='blue', label='Category A')
plt.bar(time_periods, category_B, bottom=category_A, color='red', label='Category B')
plt.bar(time_periods, category_C, bottom=np.array(category_A) + np.array(category_B), color='green', label='Category C')

# Add title and labels
plt.title("Stacked Bar Chart of Category Contributions Over Time")
plt.xlabel("Time Periods")
plt.ylabel("Value")

# Add legend
plt.legend()

# Show the plot
plt.show()
```
